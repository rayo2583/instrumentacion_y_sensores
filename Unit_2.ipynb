#Errores en las mediciones

##Error relativo y error absoluto

Al medir una cantidad con valor verdadero $ <x> $,
el valor medido $ x $ puede diferir en una pequeña cantidad $ \Delta x $. Nosotros
hablar de $ \Delta x $ como el *error absoluto* o *incertidumbre absoluta* de $ x $. A menudo, la magnitud del error o la incertidumbre es más naturalmente expresado como una fracción de lo verdadero o lo medido valor $ \bar x $. Dado que el valor verdadero $ <x>$ es
normalmente no conocido, definiremos el *error relativo* o
*incertidumbre relativa* como $ \Delta x / x $.

En la redacción científica, con frecuencia encontrará mediciones
reportadas en la forma $ x = 3.3 \pm 0.05 $. Leemos esto como $ x = 3.3 $ y
$ \Delta x = 0.05 $.

##Ejemplo 1

Un fabricante de componentes quiere construir resistencias que esten entre $1.14 k\Omega$ y $1.26 k\Omega$ y clasifica este producto com resistencias de $1.2 k\Omega$. Cual es la tolerancia de este producto? Si los valores de la resistencia se spcifican a una temperatura de $25^{\circ} \mathrm{C}$ y el material de las resistencias tiene un coeficiente térmico de $+500 \mathrm{ppm} /{ }^{\circ} \mathrm{C}$ Cual es la resistencia maxima quee puede tener uno de estos componentes a $75^{\circ} \mathrm{C}$?

###Solución

```python
import math
#Variable Declaration

#Resistance Values at 25 degree Celsius
rmax25=1.26*10**3                  #in ohm
rmin25=1.14*10**3                  #in ohm
r=1.2*10**3                        #in ohm
ppm=500.0/1000000              

#Calculations

absolute_error=rmax25-r           #in ohm  

#Tolerance value in percentage
tolerance=absolute_error/r*100    #percentage

#Resistance per degree Celsius
rperc=rmax25*ppm


#To Calculate ressistance at 75 degree celsius

dT=75-25                          #degree celsius

dR=rperc*dT                       #in ohm  

#Maximum resistance at 75 degree celsius

rmax75=rmax25+dR                  #in ohm  


#Results
print ('Tolerance of the resistance= ±',int(tolerance),'%')
print ('Maximum Resistance at 75 degree celsius=',round(rmax75/1000,4),'kilo ohm')
```

##Ejemplo 2

Calcular el máximo porcentaje de error en la suma de dos mediciones de voltaje cuando $V_{1}=100 \mathrm{~V} \pm 1 \% \text { and } V_{2}=80 \mathrm{~V} \pm 5 \%$ 

###Solución

```python
import math
#Variable Declaration

V1=100              #in V         
V2=80               #in V
p1=1.0/100          #Percentage error of V1
p2=5.0/100          #Percentage error of V2


#Calculations
V1max=V1+V1*p1      #in V
V2max=V2+V2*p2      #in V 


Emax=V1max+V2max    #in V
E=V1+V2             #in V

p=100*(Emax-E)/E    #Percentage

#Results
print ('Maximum percentage error=',round(p,1),'%')
print ('V=(',E,'V ±',round(p,1),'%)')
```

## Ejemplo 3

Calcular el máximo porcentaje de error en la suma de dos mediciones de voltaje cuando $V_{1}=100 \mathrm{~V} \pm 1 \% \text { and } V_{2}=80 \mathrm{~V} \pm 5 \%$

###Solución

```python
import math
#Variable Declaration

V1=100              #in V  
V2=80               #in V  
p1=1.0/100          #Percentage error of V1
p2=5.0/100          #Percentage error of V2


#Calculations
E=V1-V2            #in V
V1max=V1+V1*p1     #in V 
V2min=V2-V2*p2     #in V
Emax=V1max-V2min   #in V

p=100*(Emax-E)/E   #percentage

#Results
print ('Maximum percentage error= ±',int(p),'%')
print ('Voltage=(',E,'V ±',int(p), '%)')
```

## Ejemplo 4

Una resistencia de $820 \Omega$  medida con una precision de $\pm 10 \%$ conduce una corriente de $10 \mathrm{~mA}$. La corriente se midió con un amperímetro analógico en un rango de $25 \mathrm{~mA}$  con una precisión de $\pm 2 \%$ de escala completa. Calcule la potencia disipada en la resistencia y determine la precisión del resultado.

###Solución

```python
import math
#Variable Declaration
r=820                       #Resistance(ohm)
r_accuracy=10.0/100         #Accuracy of r in percentage
I=10*10**(-3)               #Current(A)
I_accuracy=2.0/100          #Accuracy of I at Full scale in percentage
Imax=25*10**(-3)            #Full scale current(A)

#Calculations
power=r*(I**2)              #in Watt

I_error=I_accuracy*Imax

I_error_percentage=100*I_error/(10*10**(-3))

Isquare_error=2*I_error_percentage

p_error=Isquare_error+(r_accuracy*100)  


#Results
print ('Power=',int(power*1000),' mW')
print ('Percentage error in power= ±',int(p_error), '%')
print ('Power=(',int(power*1000),'mW ±',int(p_error),'%)')
```

## Ejemplo 5

La precisión de cinco voltímetros digitales se verifica usando cada uno de ellos para medir un estándar $ 1.0000 \mathrm {~ V} $ de un instrumento de calibración. Las lecturas del voltímetro son las siguientes: $ V_ {1} ^ {\prime} = 1.001 \mathrm {~V}, V_{2} = 1.002, V_{3} = 0.999, V_{4} = 0.998 $ y $ V_ {5} = 1.000 $ Calcule el voltaje medido promedio y la desviación promedio.

###Solución

```python
import math

#Variable Declaration
V=[1.001,1.002,0.999,0.998,1.000]     #Voltage readings
v_average=0.0                         #Variable to hold average value
d=[0]*5                               #Array of 5 elements to hold deviation
D_average=0.0                         #Variable to hold average deviation

#Calculation
#To find average
for i in range(0,5):
    v_average=v_average+V[i]
 
v_average=v_average/5.0

#To find deviations
for i in range(0,5):
    d[i]=V[i]-v_average

#To find mean deviation    
for i in range(0,5):
    D_average=D_average+math.fabs(d[i])

D_average=D_average/5

print ('Average Deviation=',round(D_average,5),'volt')
```

## Propagación de errores en una variable

Suponga que $ x $ es el resultado de una medición y estamos calculando
una cantidad dependiente
$$
  y = f(x) \,.
$$
Sabiendo $ \Delta x $, debemos derivar $ \Delta y $, el error asociado o
incertidumbre de $ y $. Recordemos la ecuación de la recta tangente a $ f $ en el punto $ x $,
$$
  \ell(x+\Delta x) = f(x) + f'(x) \, \Delta x \,.
$$
Usamos la ecuación de la recta tangente como una aproximación (lineal) a $ f $;
cuando $ \Delta x $ no es demasiado grande, esperamos que esta aproximación sea
bueno. Por lo tanto,
$$
  \Delta y 
  = f(x+\Delta x) - f(x)
  \approx \ell(x+\Delta x) - f(x)
  = f'(x) \, \Delta x \,. 
$$
Si pensamos en $ \Delta x $ como un número positivo que especifica el
(esperada) magnitud del error, escribiremos la propagación del error
fórmula en la forma
$$
  \Delta y \approx \lvert f'(x) \rvert \, \Delta x \,.
$$



#Sensores RCIM

Los sensores resistivos, capacitivos, inductivos y magnéticos (RCIM) son los sensores más utilizados y comprenden más del 70% del mercado de sensores en la actualidad. Sus principios de fácil comprensión, características robustas, circuitos electrónicos simples, bajo costo y técnicas de diseño y fabricación maduros los han convertido en la primera opción en muchos diseños y aplicaciones de sensores. Los sensores RCIM pueden medir la presión, la humedad, la aceleración, el caudal, la radiación, la corriente eléctrica, los campos eléctricos y magnéticos o sustancias químicas y biológicas, y se utilizan en una variedad de aplicaciones, incluida la seguridad nacional, la protección del medio ambiente, la automatización del hogar, el cuidado de la salud. , la aviación, así como las industrias alimentaria, química y médica.

##Definición de un sensor RCIM 

La palabra sensor proviene del latín sentire que significa "percibir". Webster's Collegiate Dictionary define un sensor como “un dispositivo que responde a un estímulo físico (o químico) (como calor, luz, sonido, presión, magnetismo o movimiento) y transmite un impulso resultante (como se usa para medir u operar un control ). " Un sensor RCIM es un sensor que produce un cambio en la resistencia eléctrica, capacitancia, inductancia, campo eléctrico o campo magnético cuando se expone a un estímulo físico, químico o biológico. La salida podría ser un voltaje, corriente, carga eléctrica, tiempo o frecuencia para reflejar estos cambios.

##Características y terminología empleada

###Función de transferencia

Transfer.png

La función de transferencia de un sensor RCIM describe la relación entre su entrada y su salida y, a menudo, se expresa mediante una ecuación, un gráfico o una tabla. Por ejemplo, un sensor de presión capacitivo KP125 de Infineon Technologies tiene una función de transferencia de [5]
$$
V _ {\mathrm{out}} = \left (a P _ {\mathrm{in}} + b \right) V_{dd}
$$
Esta función de transferencia contiene información importante sobre el sensor: la sensibilidad $ a $, la compensación $ b $ y la relación entre la presión de entrada $ P _ {\text{in}} $ y su voltaje de salida $ V_{\text{out }} $ bajo una fuente de alimentación constante $ V_{dd} $. Esta relación insumo-producto también se puede describir gráficamente  o como una tabla.

La función de transferencia de un sensor también se utiliza para la calibración del sensor o la predicción del rendimiento. Idealmente, un sensor debería tener una función de transferencia lineal, es decir, la salida del sensor es linealmente proporcional a su entrada. En realidad, la mayoría de los sensores muestran diferentes grados de no linealidad; por lo tanto, la linealización de su función de transferencia en un punto de operación está involucrada en el diseño, modelado y control del sensor.

###Sensibilidad

La sensibilidad, la pendiente de la función de transferencia de un sensor, se define como la relación entre un pequeño cambio en la salida del sensor y un pequeño cambio en su entrada. En algunos sensores, la sensibilidad se expresa como el cambio de parámetro de entrada necesario para producir un cambio de salida estandarizado. En otros, la sensibilidad se describe mediante un cambio de salida para un cambio de entrada dado bajo la misma condición de excitación o una fuente de alimentación de voltaje constante.

Para muchos sensores, un cambio en la sensibilidad a menudo se debe a la fluctuación de la temperatura. Una curva que muestra cómo cambia la sensibilidad a medida que varía la temperatura se encuentra a menudo en la hoja de datos de un sensor. Si dicho sensor se utiliza a diferentes temperaturas, se debe realizar una calibración a cada una de estas temperaturas.

$$

###Compensación

Offset.png

Cuando la propiedad medida es cero pero el valor de salida de un sensor no es cero, entonces el sensor tiene una compensación. La compensación también se denomina compensación cero o nula, deriva cero o nula, error de compensación, polarización, compensación de CC o componente de CC. Una compensación puede describirse mediante el valor de salida de un sensor en su entrada cero  o la diferencia entre el valor de salida real del sensor y un valor de salida ideal especificado . En la parte a de la figura, la compensación del sensor está indicada por el valor b medido desde el punto de entrada cero (origen) a lo largo del eje vertical (salida). En la parte b de la Figura, la compensación b del sensor está indicada por la diferencia entre la salida real del sensor y su salida ideal en el eje vertical (salida).
La compensación se produce debido a errores de calibración, degradación del sensor a lo largo del tiempo o cambios ambientales. Entre estos, el cambio de temperatura es el factor principal que causa la deriva. Por lo general, la curva de la función de transferencia a temperatura ambiente (por ejemplo, 25 ° C) se utiliza como curva de referencia o ideal, mientras que las curvas a otras temperaturas se consideran curvas reales. El error de compensación se puede eliminar fácilmente de las salidas de un sensor restando la constante b de las salidas reales del sensor o mediante un proceso de calibración.

###Rango completo (FS) y salida de rango completo (FSO).

FSO.png

Tramo completo o escala completa (FS), o tramo para abreviar, es un término utilizado para describir el rango o limitación de medición de un sensor. Representa el rango más amplio, de mínimo a máximo, de un parámetro de entrada que puede ser medido por un sensor sin causar inexactitudes inaceptables. La salida de rango completo (FSO) es un término que se utiliza para describir el rango dinámico o el rango de salida de un sensor: la diferencia entre las salidas de un sensor medidas con el estímulo de entrada mínimo y el estímulo de entrada máximo. La Figura, tomando un sensor de presión piezorresistivo como ejemplo, ilustra FS y FSO en una función de transferencia ideal. FS es el rango entre la presión mínima Pmin y la presión máxima Pmax que puede medir el sensor; FSO es la diferencia entre los voltajes de salida Vmin y Vmax correspondientes a Pmin y Pmax, respectivamente.
Para una función de transferencia real, FSO debe incluir todas las desviaciones de la función de transferencia ideal.

El FS de un sensor puede ser unipolar (valores positivos o negativos de la medida y) o bipolar. Ejemplos de sensores con rangos unipolares son sensores de presión (p. Ej., 0 a 200 MPa) o sensores de fuerza para fuerzas de compresión (p. Ej., -30 kN a 0 N).

Los rangos bipolares pueden ser simétricos o asimétricos. Por ejemplo, un sensor de aceleración tiene un rango simétrico (p. Ej., ± 25g, donde g = 9,81 $m  s^{−2}$ es la aceleración gravitacional), mientras que un sensor de impacto tiene un rango asimétrico de −200g a + 10,000g).
Algunas hojas de datos de sensores también proporcionan sobrecarga además de los rangos FS y FSO. La sobrecarga se puede especificar como un valor (por ejemplo, 140 MPa) o como un porcentaje del intervalo (por ejemplo, 140%). Si un sensor está sobrecargado, ya no funcionará dentro de las tolerancias o precisión especificadas. Una operación de sobrecarga es la justificación para que el fabricante de un sensor rechace un reclamo de garantía.

###Precisión

La precisión se define como la desviación máxima de un valor medido por el sensor de su valor real. La precisión se puede representar en una de las siguientes formas:
Precisión absoluta: en términos de (1) parámetros medidos (por ejemplo, presión o aceleración), o (2) parámetros de salida (por ejemplo, voltaje y resistencia)
Precisión relativa: en términos de (1) un porcentaje del error máximo de medición frente al valor real:
$$
\frac{\operatorname{Max} \{\text{valor medido} - \text{valor verdadero} \}} {\text{valor verdadero}} \times 100 \%
$$
o (2) un porcentaje del error de medición máximo frente al intervalo completo:
$$
\frac{\text{Max} \{\text{valor medido} - \text{valor verdadero}\} \times 100\%} {\text{intervalo completo}}
$$
Por ejemplo, la precisión de un sensor de presión piezorresistivo puede especificarse como $ \pm 0.5 \mathrm{kPa} $ (en términos del parámetro medido), $ \pm 0.05 \Omega $ (en términos del valor de resistencia de salida) o $ \pm 0.5 \% $ (precisión relativa a su valor real).

Muchos factores pueden afectar la precisión de un sensor, incluida la fluctuación de temperatura, linealidad, histéresis, repetibilidad, estabilidad, compensación cero, error de conversión A/D (analógico a digital) y resolución de pantalla. La precisión general de un sensor se puede expresar mediante la siguiente ecuación:
$$
\text{Precisión general} = \pm \sqrt {e_{1}^{2} + e_{2}^{2} + e_{3}^{2} + \cdots}
$$
donde $ e_{1}, e_{2}, e_{3}, \ldots $ representa cada componente que contribuye a la precisión.

####Ejemplo

Un termopar tipo T tiene un error de medición de $ \pm 1.1 ^ {\circ} \mathrm{F} $. Su transmisor, conversión A / D, resolución de pantalla y efecto de temperatura del cable es $ \pm 0.8 ^ {\circ} \mathrm{F} $. ¿Cuál es la precisión general del sensor?

#####Solución

Precisión general $=\pm \sqrt{\left(1.1^{\circ} \mathrm{F}\right)^{2}+\left(0.8^{\circ} \mathrm{F}\right)^{2}}=\pm 1.36^{\circ} \mathrm{F}$

###Histéresis

Histeresis.png

Un sensor puede producir diferentes salidas cuando mide la misma cantidad dependiendo de la “dirección” en la que se ha aproximado al valor. La diferencia máxima en términos de la cantidad medida se define como *error de histéresis*, denotado como $\delta_h$, o la desviación máxima en valor entre el ciclo creciente y decreciente medido en el mismo punto de entrada (ver Figura).

$\Delta_h$, o la desviación máxima en valor entre el ciclo creciente y decreciente medido en el mismo punto de entrada (ver Figura).
El *porcentaje de error de histéresis* se determina tomando la desviación máxima y dividiéndola por FS:

$$
\delta_{h}=\frac{\operatorname{Max}\left(\Delta_{h}\right)}{\mathrm{FS}} \times 100 \%
$$

####Ejemplo

Un sensor de desplazamiento potenciométrico produce una diferencia máxima de 20 mV al medir la distancia de un objeto que pasa por un cierto punto: una medida se toma cuando el objeto se mueve de izquierda a derecha hacia el punto, y la otra se toma cuando el el objeto se mueve de derecha a izquierda hasta el punto. Encuentre el error de histéresis si la sensibilidad del sensor es de 10 mV / mm.

#####Solución

Dado que el error de histéresis se define como la desviación máxima en términos de la cantidad medida, que es el desplazamiento en este caso, necesitamos convertir la salida de $ 20 \mathrm{mV} $ en la entrada de desplazamiento en milímetros:

Error de Histeresis $(\mathrm{mm})=\frac{\text { maximum output voltage difference }}{\text { sensitivity }}=\frac{20 \mathrm{mV}}{10 \mathrm{mV} \cdot \mathrm{mm}^{-1}}=2 \mathrm{~mm}$

###No-linealidad

La no linealidad describe la "rectitud" de la función de transferencia de un sensor. Una no linealidad se puede expresar mediante la desviación máxima de una función de transferencia real de su línea recta de mejor ajuste, $ \Delta_{l} $, o en porcentaje de $\Delta_{l} $ en todo el intervalo:
$$
\text{No linealidad} (\%) = \frac{\operatorname{Max} \left\{\Delta_{l} \right\}} {\mathrm{FS}}\times 100 \%
$$
La no linealidad no debe confundirse con la precisión. Este último indica qué tan cerca está un valor medido de su valor real, pero no qué tan directa es la función de transferencia.

####Ejemplo

El error de linealidad de un sensor de desplazamiento capacitivo en su peor punto es de 0,01 mm. Si el rango de escala completa de su función de transferencia es 1 mm, ¿cuál es su no linealidad expresada como porcentaje?

#####Solución

$$
\text { Nonlinearity }(\%)=\frac{\operatorname{Max}\left\{\Delta_{l}\right\}}{\mathrm{FS}} \times 100 \%=\frac{0.01 \mathrm{~mm}}{1 \mathrm{~mm}} \times 100 \%=1 \%
$$

###Razón señal a ruido

Todos los sensores generan ruido además de sus señales de salida. El ruido del sensor se puede definir como cualquier desviación o fluctuación de un valor esperado. Para describir estas fluctuaciones, un promedio simple no tiene sentido ya que el promedio de las variaciones aleatorias es cero. En su lugar, se usa la raíz cuadrada media (RMS) de las desviaciones de $ V _ {\text{average}} $ durante un intervalo de tiempo $ T $:
$$
\sqrt{V_{n}^{2}} = \sqrt{\frac{1}{T} \int_{0}^{T} \left(V-V_{\text{average}} \right)^{2} dt}
$$
donde $ V_ {n} ^ {2} $ es el ruido cuadrático medio en voltaje. $ \sqrt {V_ {n} ^ {2}} $ es el ruido RMS en voltaje.
La medida en que el ruido se vuelve significativo en un proceso de detección o medición depende de la amplitud relativa de la señal de interés al valor del ruido no deseado, llamado relación señal / ruido (relación SNR o relación S / N). Si el valor del ruido es pequeño en comparación con el nivel de la señal, entonces la SNR es grande y el ruido deja de ser importante. En el diseño de sensores, siempre se intenta maximizar la SNR.

La SNR se puede calcular mediante
$$
\mathrm{SNR} = \frac{\bar{P}_{s}} {\bar{P}_{n}} = \frac{V_{s}^{2}} {V_{n} ^ { 2}}
$$
o en decibelios (dB)
$$
\mathrm{SNR} = 10 \log_{10}\left(\frac{V_{s}^{2}} {V_{n}^{2}}\right) = 20\log_{10}\left(\frac{V_{s}}{V_{n}}\right)
$$
donde $\bar{P}_{s} $ y $ \bar{P}_{n} $ son la potencia promedio de la señal y el ruido, respectivamente, y $ V_{s} $ y $ V_{n} $ son los voltajes RMS de la señal y el ruido, respectivamente.
Otra medida cuantitativa del ruido es el factor de ruido, $ F_{n} $, definido como
$$
F_{n} = \frac{\text{SNR en la entrada}} {\text{SNR en la salida}}
$$
El factor de ruido $ F_{n} $ especifica a cuánto ruido adicional contribuye el sensor además del ruido ya recibido de la fuente. Idealmente, $ F_{n}= 1 $

####Ejemplo

El voltaje de salida esperado de un sensor es de 100 mV. Sin embargo, un voltaje de ruido de 18 μV se superpone a su señal de salida. Encuentra la SNR del sensor.

#####Solución

$$
\text { SNR }=\frac{\left(100 \times 10^{-3} \mathrm{~V}\right)^{2}}{\left(18 \times 10^{-6} \mathrm{~V}\right)^{2}}=3.09 \times 10^{7}
$$
ó
$$
\mathrm{SNR}=20 \log _{10}\left(\frac{100 \times 10^{-3}}{18 \times 10^{-6}}\right)=74.90 \mathrm{~dB}
$$

###Resolución

La resolución de un sensor especifica el cambio más pequeño de parámetro de entrada que se puede detectar y reflejar en la salida del sensor. En algunos sensores, cuando un parámetro de entrada cambia continuamente, es posible que la salida del sensor no cambie suavemente, sino en pequeños pasos. Esto suele ocurrir en sensores potenciométricos y detectores de presencia de infrarrojos con máscaras de rejilla.
La resolución se puede describir en términos absolutos o relativos (en porcentaje de FS). Por ejemplo, un sensor magnético HMC100 de Honeywell tiene una resolución de 27 µgauss (microgauss) [6] en términos de un valor absoluto, lo que significa que el sensor puede detectar un cambio mínimo de 27 µgauss; un sensor de temperatura digital ADT7301 de Analog Devices tiene una resolución de 0.03 ° C / LSB (bit menos significativo) [7], lo que significa que un bit puede distinguir un cambio de temperatura de 0.03 ° C; y un sensor de humedad SHT1 de SENSIRION tiene una resolución de 0.05% de humedad relativa (RH) [8], lo que significa que la resolución del sensor es 0.05% de su rango.
La resolución de un sensor debe ser mayor que la precisión requerida en la medición. Por ejemplo, si una medición requiere una precisión de 0,02 μm, la resolución del sensor debe ser mejor que 0,02 μm.
Los factores que afectan la resolución varían de un sensor a otro. Para la mayoría de los sensores capacitivos, el principal factor que afecta es el ruido eléctrico. Tomando un sensor de desplazamiento capacitivo como ejemplo, aunque la distancia entre el sensor y el objetivo es constante, la salida de voltaje seguirá fluctuando ligeramente debido al ruido "blanco" del sistema. Suponga que no hay acondicionamiento de señal, no se puede detectar un cambio en la salida de voltaje que sea menor que el valor de voltaje pico a pico del ruido. Debido a esto, la mayoría de las resoluciones de los sensores capacitivos se evalúan por el valor pico a pico del ruido dividido por la sensibilidad:
$$
\text{Resolución = valor pico a pico de ruido / sensibilidad}
$$

####Ejemplo

El valor pico a pico de ruido de un sensor de desplazamiento capacitivo es 0.01 V. Si la sensibilidad del sensor es 10 V ⋅ mm − 1, ¿cuál es su resolución?

#####Solución

$$
\text { Resolution }=\text { peak-to-peak value of noise/sensitivity }=0.01 \mathrm{~V} /\left(10 \mathrm{~V} \cdot \mathrm{mm}^{-1}\right)=1 \mu \mathrm{m}
$$

###Error de precisión y repetibilidad

repeability.png

La precisión se refiere al grado de repetibilidad o reproducibilidad de un sensor. Es decir, si se mide exactamente el mismo valor varias veces, un sensor ideal produciría exactamente la misma salida cada vez. Pero los sensores en realidad emiten un rango de valores distribuidos de alguna manera en relación con el valor correcto real. La precisión se puede expresar matemáticamente como
$$
\text { Precision }=1-\left|\frac{x_{n}-\bar{x}}{\bar{x}}\right|
$$
donde $ x_ {n} $ es el valor de la $ n $ ésima medida y $ \bar {x} $ es el promedio del conjunto de $ n $ medidas.

El error de repetibilidad describe la incapacidad de un sensor para reconocer el mismo valor en condiciones idénticas. La figura $ 1.7 $ muestra las dos corridas de la misma medición realizadas por el sensor $ A $ y el sensor $ B $. Aparentemente, el sensor $ B $ tiene un error de repetibilidad menor que el sensor $ A $. Por lo general, el error de repetibilidad se expresa en un porcentaje de la escala completa.
$$
\delta_{r}=\frac{\operatorname{Max}\left(\Delta_{r}\right)}{\mathrm{FS}} \times 100 \%
$$
Por ejemplo, un sensor de fuerza A201 de Tekscan normalmente tiene un error de repetibilidad de $ \pm 2.5 \% $ FS. El error de repetibilidad puede ser causado por ruido térmico, cargas acumuladas, plasticidad del material y fricción.

###Error de calibración

Muchos sensores requieren calibración antes de su uso. Para algunos sensores, la calibración consiste en determinar los coeficientes en sus funciones de transferencia (desplazamiento, sensibilidad o ambos); para otros, calibración significa convertir la salida eléctrica del sensor (por ejemplo, voltaje) a un valor medido (por ejemplo, temperatura). Aunque los procedimientos de calibración detallados varían de un sensor a otro, el procedimiento de calibración general implica tomar una serie de mediciones en todo el rango operativo y comparar los valores medidos con los valores reales (o trazar la curva de entrada-salida del sensor contra una curva estándar / de referencia), y luego ajustar los parámetros para que coincidan con los valores reales (o la curva estándar). Los valores reales (o la curva estándar) pueden (1) obtenerse utilizando un sensor más preciso como referencia de calibración o $ (2) $ proporcionados por el fabricante del sensor. Los sensores inteligentes o avanzados pueden tener funciones de autocalibración y adaptación para ajustarse a diferentes condiciones de operación.

Los siguientes ejemplos demuestran los procedimientos de calibración de un punto y dos puntos del sensor de temperatura lineal LM335 de National Semiconductor. Su función de transferencia especificada por el fabricante es
$$
V_{\mathrm{out}}=\left(T_{n}+273.15 \mathrm{~K}\right)(10 \mathrm{mV} / \mathrm{K})
$$
donde $ V _{\text {out}} $ es su voltaje de salida proporcional a la temperatura absoluta que se medirá en su rango útil $\left(-40^{\circ} \mathrm {C} \right. $ a $ \left. + 100 ^ {\circ} \mathrm {C} \right) $ y $ T_ {n} $ (en $\left.{}^{\circ}\mathrm{C}\right) $ es el nominal temperatura.

El error de calibración es la cantidad de inexactitud permitida cuando se calibra un sensor. Este error es a menudo de naturaleza sistemática, lo que significa que puede ocurrir en todas las funciones de transferencia reales. Puede ser constante o puede variar en el rango de medición, dependiendo del tipo de error en la calibración.

###Tiempo de respuesta y ancho de banda

overdamped_response.png

Estrictamente hablando, los sensores no responden inmediatamente a un estímulo de entrada. Más bien, tienen un tiempo de respuesta finito o un tiempo de subida $ t_{r} $ ante un cambio instantáneo de estímulo. El tiempo de respuesta puede definirse como el "tiempo necesario para que la salida de un sensor cambie de su estado anterior a un valor establecido final dentro de una banda de tolerancia del nuevo valor correcto" [11]. Por lo general, $ t_{r} $ se determina como el tiempo necesario para que la señal de un sensor cambie de un valor bajo especificado a un valor alto especificado de su respuesta al escalón. Normalmente, estos valores son $ 10 \% $ y $ 90 \% $ de su valor final para sobreamortiguado, $ 5 \% $ y $ 95 \% $ de su valor final para críticamente amortiguado, y $ 0 \% $ y $ 100 \% $ de su valor final para la respuesta escalonada subamortiguada. El sensor de fuerza A201 de FlexiForce tiene un tiempo de respuesta de menos de $ 5 \mu \mathrm {s} $ [9], que es típico para la mayoría de los sensores. La constante de tiempo $ \tau $ representa el tiempo que tarda la respuesta al escalón del sensor en alcanzar $ 1-1/e\approx 0,632 $ o $ 63,2 \% $ de su valor final. La figura $ 1.8 $ indica la respuesta / tiempo de subida de un sensor $ t_{r} $ y su constante de tiempo $ \tau $ bajo un estímulo escalonado positivo en un caso sobreamortiguado.

El sobreimpulso es la salida máxima de un sensor sobre su salida estable final en respuesta a una entrada escalonada (consulte la Figura 1.9). En este caso, $ t_{r} $ es el tiempo necesario para que la salida de un sensor aumente de $ 0\% $ a $ 100 \% $ de su valor de estado estable; el tiempo pico $ t_{p} $ es el tiempo necesario para que el sensor alcance el primer pico de su respuesta escalonada; El tiempo de estabilización $ t_{s} $ es el tiempo necesario para que la salida del sensor alcance y se mantenga dentro de $ \pm 2 \% $ de su valor final. El error de estado estable es el error entre el valor de salida real y el deseado.
El sobreimpulso en porcentaje (llamado sobreimpulso porcentual) se puede obtener mediante
$$
\text { P.O. }(\text { percentage overshoot })=\frac{\text { maximum peak value }-\text { final value }}{\text { final value }} \times 100 \%
$$

###Vida útil del sensor

Todos los sensores tienen una vida útil limitada, indicada por la vida útil / operativa, la vida útil cíclica, la clasificación continua, la clasificación intermitente, la vida útil en almacenamiento o la fecha de vencimiento. Varios factores pueden afectar la vida de un sensor: su tipo, diseño, material, frecuencia y duración de uso, niveles de concentración que se medirán, proceso de fabricación, esfuerzos de mantenimiento, aplicación, almacenamiento y condiciones ambientales (por ejemplo, temperatura, humedad) [14] . Los sensores que consumen materiales internos durante el proceso de detección (por ejemplo, ciertos sensores de glucosa u oxígeno) se pueden usar solo una o pocas veces. Algunos sensores de gas (por ejemplo, los sensores de CO y H2S de Biosystems), aunque no son consuntivos, solo tienen un límite de vida útil de 2 a 4 años debido a factores como la evaporación (secado), las fugas y la contaminación del catalizador. Los sensores de oxígeno utilizados en motores de automóviles que se construyeron antes de 1995 tienen una vida útil de 50,000 millas, mientras que los diseños más nuevos (1996 y posteriores) solo necesitan ser reemplazados cada 100,000 millas. Un sensor de dióxido de azufre expuesto a una concentración de 5 ppm (partes por millón) continuamente puede durar 10,000 horas o más de un año; si la concentración es inferior a 5 ppm, se prolongará la vida útil. La mayoría de los sensores mecánicos tienen una vida útil prolongada en condiciones normales de funcionamiento. Por ejemplo, los sensores de carga de FlexiForce tienen una vida útil de un millón de ciclos de carga con cargas de menos de 50 lb (vida útil cíclica). Muchos sensores de temperatura tienen una vida útil de más de 10 años.
El envejecimiento también afecta la precisión de algunos sensores y hace que los sensores pierdan sensibilidad lentamente con el tiempo. Por ejemplo, ciertos tipos de sensores de temperatura reducen su sensibilidad en 0,1 ° C cada año. La precisión de la mayoría de los químicos y biosensores depende de sus edades. Además, el manejo brusco de un sensor también podría acortar su vida útil. Un sensor que se instala y quita repetidamente tendrá una vida más corta que un sensor que se instala y se deja en su lugar. La mejor manera de prolongar la vida útil de los sensores es almacenarlos correctamente, probarlos y verificar su precisión con regularidad y volver a calibrarlos cuando sea necesario. Los sensores más avanzados de hoy están equipados con sistemas de monitoreo de la vida útil de los sensores para recordar a los usuarios cuándo estos sensores deben ser reemplazados. Los sensores deben reemplazarse cuando ya no se puedan calibrar o poner a cero fácilmente.

#Sensores resistivos

La resistencia eléctrica es la propiedad eléctrica más fácil de medir con precisión en un amplio rango a un costo moderado. Los sensores resistivos tienen muchas características deseables: confiabilidad, resolución ajustable, construcción simple y facilidad de mantenimiento, que los han convertido en la opción preferida en el diseño y la aplicación de sensores. Según sus principios de detección, los sensores resistivos se clasifican como:




*   **Sensores potenciométricos**: cambio de resistencia debido al cambio de posición lineal o angular
*   **Sensores de temperatura resistivos**: cambio de resistencia causado por la variación de temperatura (efecto termorresistivo)
*  **Sensores fotorresistivos**: disminución de la resistencia cuando la luz incide en un material fotoconductor (efecto fotorresistivo)
*  **Sensores piezorresistivos**: cambio de resistencia cuando se aplica una fuerza a un conductor piezorresistivo (efecto piezorresistivo)
*  **Sensores magnetorresistivos**: cambio de resistencia en presencia de un campo magnético externo (efecto magnetorresistivo)
*  **Sensores quimiorresistivos**: cambio de conductividad en un material o solución debido a reacciones químicas que alteran el número de electrones o la concentración de iones.
* **Sensores biorresistivos**: cambio de biorresistencia en proteínas o células inducido por variaciones estructurales e interacciones biológicas. 

##Sensores potenciométricos

### Principio de sensado

Los sensores potenciométricos están diseñados en base a la ecuación para la resistencia de un conductor. La resistencia de un conductor $ R $ (en ohmios, $ \Omega) $ es una función de la resistividad del material conductor $ \rho $ (en ohmímetro, $ \Omega \cdot \mathrm{m}) $, su longitud $ l $ (en metros, $ \mathrm {m}) $, y su área de sección transversal $ A $ (en metros cuadrados, $ \mathrm {m} ^ {2} $):
$$
R = \rho \frac{l}{A}
$$
Tecnologías de sensores resistivos, capacitivos, inductivos y magnéticos
Aunque cualquier cambio en $ 1, A $ y $ \rho $ provocará un cambio en la resistencia, los sensores potenciométricos (también llamados potenciómetros o potenciómetros para abreviar) a menudo se diseñan variando la longitud $ l $ solo por simplicidad y para ahorrar costos. Algunos sensores quimiorresistivos están diseñados en función del cambio de resistividad de los materiales $ \rho $ causado por reacciones químicas.

La resistencia de un potenciómetro se puede evaluar usando la Ley de Ohm aplicando una corriente eléctrica $ I $ (en amperios, $ \mathrm {A} $) y midiendo el voltaje $ V $ (en voltios, $ \mathrm{V}) $ a través del potenciómetro:
$$
R = \frac {V} {I}
$$

###Diseño

####Potenciometros Lineales

Los componentes principales de un potenciómetro lineal incluyen tres terminales (entrada de energía, tierra y salida de señal del sensor), una resistencia fija, un limpiaparabrisas, una pista deslizante y la carcasa. Los terminales suelen estar chapados en oro para evitar la corrosión o el deslustre. La resistencia generalmente está hecha de materiales conductores de alta calidad (por ejemplo, cobre) con un coeficiente de temperatura más bajo para estabilidad y larga vida. El limpiaparabrisas utiliza aleación o metal de alta calidad (por ejemplo, una aleación de platino) para una vida útil prolongada y un nivel de ruido reducido. Algunos limpiaparabrisas incluso tienen una forma de varios dedos para evitar intermitencias cuando se utilizan en entornos con fuertes impactos o vibraciones. El limpiaparabrisas está conectado al objeto en movimiento que se está midiendo a través de roscas, chaflanes o retornos de resorte. Algunos sensores potenciométricos tienen otros componentes, como rieles de guía (para mejorar la estabilidad durante el deslizamiento del limpiaparabrisas) y arandelas onduladas para control antibacklash. La carcasa mantiene los componentes en su lugar y protege el sensor de entornos hostiles.
Las principales consideraciones en el diseño de un sensor potenciométrico lineal incluyen
* La longitud de la carrera a medir
* Potencia nominal y valor de resistencia
* Las limitaciones de espacio
* La calidad del elemento conductor.
* Los medios para conectar el sensor al dispositivo móvil que se está midiendo.

####Potenciometros Rotativos

Hay dos diseños diferentes de potenciómetros giratorios: de una vuelta y multivuelta. Un potenciómetro de una sola vuelta gira menos de una revolución completa (es decir, $ < 360$ °) para alcanzar su rango de resistencia completo. A menudo se utiliza como sensor de posición angular. Un potenciómetro de múltiples vueltas gira más de una revolución (por ejemplo, 5, 10, 20 o 25 vueltas) para alcanzar su rango de resistencia completo, y a menudo se usa cuando se requiere un ajuste de mayor resolución o múltiples revoluciones. Los potenciómetros multivuelta son más caros, pero muy estables con alta precisión. Al igual que los potenciómetros lineales, los potenciómetros de una sola vuelta y de varias vueltas tienen tres terminales, una resistencia fija, un limpiaparabrisas giratorio, un eje y la carcasa. La resistencia de alambre enrollado en una olla giratoria a menudo usa un material que tiene un coeficiente de temperatura más bajo para mejorar la estabilidad. A veces, se moldea un plástico conductor sobre la resistencia de alambre enrollado para reducir la inductancia y proteger la resistencia.
Las principales consideraciones en un diseño de potenciómetro giratorio de una o varias vueltas son
* Ángulo o rotación a medir
* Potencia nominal y valor de resistencia
* Resolución. Elija un diseño de una sola vuelta si las resoluciones superiores  al 10% son
aceptables (menos costoso). Elija un diseño multivuelta si se requiere una resolución más alta (por ejemplo, 1%: control más fino, mejor linealidad y estabilidad)

##Sensores resistivos de temperatura

###Efecto termoresistivo

####Efecto termoresistivo en metales

La resistencia eléctrica de un conductor metálico aumenta a medida que aumenta la temperatura. Esto se debe a que la conductividad eléctrica de un metal depende del movimiento de los electrones a través de su red cristalina. Debido a la excitación térmica, la vibración de los electrones aumenta, lo que ralentiza el movimiento de los electrones, lo que hace que aumente la resistencia. La relación entre la resistencia $ R $ y la temperatura $ T $ se puede expresar mediante una ecuación polinomial:
$$
R_{T} = R_{0} \left[1 + A \left(T-T_ {0} \right) + B \left(T-T_ {0} \right) ^ {2} + C \left( T-T_ {0} \right) ^ {3} + \cdots \right]
$$
Su versión simplificada es
$$
R_ {T} = R_ {0} \left[1 + A \left(T-T_{0} \right)\right]
$$
donde $ R_ {0} $ es la resistencia a la temperatura de referencia $ T_ {0} $ (generalmente $ 0 ^ {\circ} \mathrm{C}, 20 ^ {\circ} \mathrm{C} $ o $ \left.25^{\circ} \mathrm{C} \right); A, B, C, \ldots $ son coeficientes de temperatura dependientes del material (en $ \left.\Omega \cdot \Omega^{- 1}\cdot {}^{\circ} \mathrm{C}^{-1} \right) $ Los metales tienen coeficientes de temperatura positivos (PTC), porque su resistencia aumenta a medida que aumenta la temperatura. Todos los dispositivos de temperatura de resistencia (RTD), hechos de metales, son sensores PTC. El coeficiente de temperatura para todos los metales puros es del orden de $ 0.003-0.007 \Omega \cdot \Omega^{- 1} \cdot{}^ {\circ} \mathrm{C}^{- 1}. $ Coeficientes de temperatura $ A $ para metales comunes se enumeran en la Tabla.

####Efecto termoresistivo en semiconductores

En los materiales semiconductores, los electrones de valencia están unidos en enlaces covalentes con sus vecinos. A medida que aumenta la temperatura, la vibración térmica de los átomos rompe algunos de estos enlaces y libera electrones. Estos electrones "libres" pueden moverse a través del material bajo campos eléctricos aplicados y el material parece tener una resistencia menor. Por tanto, la resistencia eléctrica $ R $ de los materiales semiconductores disminuye a medida que aumenta la temperatura $ T $. La relación entre $ R $ y $ T $ es exponencial:
$$
R_ {T} = R_ {0} e ^ {\left [\beta \left (\frac{1} {T} - \frac{1} {T_{0}} \right) \right]}
$$
donde $ R_ {0} $ es la resistencia a la temperatura de referencia $ T_{0} $ (en kelvin, $ \left. \mathrm{K} \right) $, normalmente $ 298 \mathrm{~ K} \left(25 ^ {\circ} \mathrm{C} \right) $, y $ \beta $ es el coeficiente de temperatura (en $ \left. \mathrm{K} \right) $ del material. Dado que la resistencia disminuye a medida que aumenta la temperatura, $ \beta $ es un coeficiente de temperatura negativo (NTC). La mayoría de los termistores (una contracción de las palabras térmico y resistor), hechos de materiales semiconductores, son sensores NTC. La ley de Wiedemann–Franz para metales 

La ley empírica de Wiedemann y Franz, propuesta en 1853, establece que la relación de la conductividad térmica $ \kappa $ (en $ \mathrm{W} \cdot \mathrm{m} ^ {- 1} \cdot \mathrm{K }^{- 1} $) a la conductividad eléctrica $ \sigma_{\mathrm{p}} $ (en $ \mathrm{S} \cdot \mathrm{m}^{- 1} $) de un metal es proporcional a su temperatura absoluta $ T $ (en $ \mathrm{K}$):
$$
\kappa / \sigma_{\rho} = L T
$$
donde $ L $ es el Número de Lorenz (en W. $ \left. \Omega.\mathrm{K}^{- 2} \right) $. Esta relación se basa en el hecho de que tanto el transporte de calor como el de electricidad se ven afectados por los electrones libres del metal. La conductividad térmica aumenta con la velocidad promedio de las partículas, ya que esto aumenta el transporte de energía hacia adelante. La conductividad eléctrica, por otro lado, disminuye cuando aumenta la velocidad de las partículas porque las colisiones desvían los electrones del transporte de carga hacia adelante.

El número de Lorenz se puede determinar teóricamente mediante la siguiente ecuación:
$$
L = \frac {\kappa} {\sigma _ {\rho} T} = \frac{\pi^{2} k_{B}^{2}} {3q^{2}} = 2,44 \times 10^{-8} \mathrm{~ W} \cdot \Omega \cdot \mathrm{K}^{- 2}
$$
donde $ k_{B} $ es la constante de Boltzmann $ \left (1.38065\times 10 ^{- 23} \mathrm{~ J} \cdot \mathrm{K} ^ {- 1} \right) $ y $ q $ es la carga del electrón $\left (1.602 \times10^{-19}\mathrm{C}\right) $. El número de Lorenz también se puede determinar experimentalmente. En la tabla $ 2.5 $ se muestran algunos números de Lorenz obtenidos experimentalmente para diferentes metales.

#####Mediciones RTD

Un RTD es un dispositivo pasivo que requiere el paso de una corriente para producir un voltaje medible. Si la corriente de excitación que pasa a través de un RTD es $ I _ {\mathrm{ex}} $, y el voltaje de salida a través del RTD es $ V _ {\text{out}} $, entonces la temperatura medida $ T \left (\right . $ en $ \left.{}^{\circ} \mathrm {C}\right) $ se puede obtener
$$
T=\frac{2\left(V_{\text {out }}-I_{\mathrm{ex}} R_{0}\right)}{I_{\mathrm{ex}} R_{0}\left[A+\sqrt{A^{2}+4 B\left(V_{0}-I_{\mathrm{ex}} R_{0}\right) /\left(I_{\mathrm{ex}} R_{0}\right)}\right]}
$$
donde $ R_ {0} $ es la resistencia del RTD en $ 0 ^ {\circ} \mathrm {C}; A $ y $ B (C = 0 $ en este caso cuando $ \left.T> 0 ^ {\circ} \mathrm{C} \right) $ son los coeficientes Callendar-Van Dusen. La corriente de excitación también hace que el RTD se caliente internamente (autocalentamiento), lo que puede resultar en un error de medición. El autocalentamiento generalmente se especifica como la cantidad de energía que elevará la temperatura del RTD en $ 1 ^ {\circ} \mathrm {C} $ (en $ \left.\mathrm {mW} \cdot {} ^ {\circ} \mathrm{C} ^ {- 1} \right) $. Para minimizar el error provocado por el autocalentamiento, se debe utilizar la corriente de excitación más pequeña posible (1 mA o menos) en la medición. La cantidad de autocalentamiento también depende en gran medida del medio en el que se sumerge el RTD. Por ejemplo, un RTD puede autocalentarse hasta 100 veces más en aire en calma que en agua en movimiento. Además, la temperatura en el RTD debe ser uniforme. De lo contrario, aparecerá un error, que está directamente relacionado con la diferencia de temperatura entre los cables del RTD.

######Ejemplo

Se pasa una corriente de excitación de $ 0.15 \mathrm {~ mA} $ a través de un RTD Pt100 fabricado según la norma DIN 43760. Si la salida de voltaje es $ 33 \mathrm {mV} $, ¿cuál es la temperatura medida (suponga $ \left.T> 0 ^ {\circ} \mathrm {C} \right) $?

######Solución

De la tabla 2.3, $ A = 3.9080 \times 10 ^ {- 3} $ y $ B = -5.8019 \ times 10 ^ {- 7} $, y $ C = 0 $, por lo tanto
$$
\begin{aligned}
T &=\frac{2\left(V_{\text {out }}-I_{\mathrm{ex}} R_{0}\right)}{I_{\mathrm{ex}} R_{0}\left[A+\sqrt{A^{2}+4 B\left(V_{0}-I_{\mathrm{ex}} R_{0}\right) /\left(I_{\mathrm{ex}} R_{0}\right)}\right]} \\
=& \frac{2\left(0.033-0.15 \times 10^{-3} \times 100\right)}{} &{\left(0.15 \times 10^{-3} \times 100\right)\left[3.9080 \times 10^{-3}+\sqrt{\left(3.9080 \times 10^{-3}\right)^{2}+4\left(5.8019 \times 10^{-7}\right)(0.033}{\left.-0.15 \times 10^{-3} \times 100\right) /\left(0.15 \times 10^{-3} \times 100\right)}\right]} \\
=& 322.85^{\circ} \mathrm{C}
\end{aligned}
$$
Los RTD pueden ser difíciles de medir debido a su resistencia relativamente baja (p. Ej., $ 100 \Omega $), que cambia solo ligeramente con la temperatura $ \left (< 0.4 \Omega \cdot {} ^ {\circ} \mathrm {C} ^ {-1} \right) $. Para medir con precisión estos pequeños cambios en la resistencia, se deben utilizar configuraciones especiales para minimizar los errores de la resistencia del cable conductor. La sección $ 6.5 .2 $ del capítulo 6 trata sobre cómo cablear correctamente el RTD para minimizar el error de medición.

```python
import math
V_out = 0.033  #Volts
V_0 = 0.033  #Volts
I_ex  = 0.15*10**-3 #
R_0 = 100 #
A = 3.9080*10**-3
B = -5.8019*10**-7
C = 0
term1 = math.sqrt(A**2 + 4.0*B*(V_0-I_ex*R_0)/(I_ex*R_0))
T = 2.0*(V_out-I_ex*R_0)/(I_ex*R_0*(A+term1))
print("{:}".format(T))
```

#####Termistores

2-14.png

Los termistores son del tipo NTC, cuya resistencia disminuye al aumentar la temperatura, o del tipo PTC, cuya resistencia aumenta al aumentar la temperatura. Cada tipo tiene características únicas y ventajas distintas.
Los termistores NTC se utilizan más comúnmente que el tipo PTC, especialmente en aplicaciones de medición de temperatura. Los termistores NTC se fabrican con tecnología de cerámica básica y materiales semiconductores de óxidos metálicos (por ejemplo, óxidos de manganeso, níquel, cobalto, hierro, cobre y titanio). En algunos termistores, la disminución de la resistencia es tan grande como un 6% por cada 1 ° C de aumento de temperatura, aunque los cambios del 1% son más típicos. Los termistores NTC pueden proporcionar una buena precisión y resolución al medir temperaturas entre −100 ° C y + 300 ° C. Si se inserta en un puente de Wheatstone, un termistor puede detectar cambios de temperatura tan pequeños como ± 0,005 ° C.
Los termistores PTC se pueden dividir en dos categorías: resistencias de silicio térmicamente sensibles (silistores) y termistores PTC de conmutación. Los silistores exhiben un PTC bastante uniforme (aproximadamente + 0.0077 ° C − 1) en la mayor parte de su rango operativo, pero también pueden exhibir una región NTC a temperaturas superiores a 150 ° C. Este tipo de termistores se utiliza a menudo para la compensación de temperatura de dispositivos semiconductores de silicio en el rango de -60 ° C a + 150 ° C. Los termistores de conmutación PTC están hechos de materiales cerámicos policristalinos que normalmente son altamente resistivos pero se vuelven semiconductores al agregar dopantes. A menudo se fabrican utilizando composiciones de bario, plomo y estroncio con aditivos como itrio, manganeso, tantalio y sílice. Las curvas R – T de los termistores PTC de conmutación exhiben regiones NTC muy pequeñas hasta que alcanzan una temperatura crítica Tc: temperatura “Curie”, “Interruptor” o “Transición”. Después de Tc, la curva muestra una resistencia a PTC que aumenta rápidamente. Estos cambios de resistencia pueden ser de varios órdenes de magnitud dentro de un intervalo de temperatura de unos pocos grados. La Figura 2.14 ilustra las curvas R – T de los termistores de silistor y de conmutación PTC [13].
La mayoría de las aplicaciones de los termistores PTC se basan en la condición de autocalentamiento de estado estable (característica de voltaje-corriente) o en la condición de autocalentamiento dinámico (característica de corriente-tiempo) o una combinación de ambas. Por ejemplo, el aumento espectacular de la resistencia de un PTC a la temperatura de transición y por encima de ella lo hace ideal para la protección contra sobrecorriente (fusibles reajustables). Si ocurre una condición de sobrecorriente, el termistor se autocalentará más allá de la temperatura de transición y su resistencia aumentará dramáticamente. Esto hace que se reduzca la corriente en el circuito general. Las características clave de los termistores NTC y PTC incluyen :

* **Características de la temperatura de resistencia (R – T):** Describa cómo cambia la resistencia a medida que cambia la temperatura en un termistor. La mayoría de los fabricantes de termistor proporcionan tablas, curvas R – T y coeficientes para sus productos de termistor.
* **Tolerancia de resistencia:** especifica las tolerancias estándar disponibles para cada tipo de termistor.
*Tolerancia beta: es la tolerancia del valor beta de un termistor, determinada por la composición y estructura de los óxidos metálicos que se utilizan en el sensor. Para termistores tipo perla, las tolerancias beta son del orden de ± 1% a ± 5%; para el tipo de contacto de superficie metalizada, las tolerancias beta oscilan entre ± 0,5% y ± 3%. Capacidad calorífica: Es el producto del calor específico y la masa del termisor. Representa la cantidad de calor necesaria para producir un cambio de 1 ° C en la temperatura corporal del termistor.
* **Constante de disipación:** es la relación entre el cambio en la potencia aplicada a un termistor y el cambio resultante en la temperatura corporal debido al autocalentamiento. Se ve afectado por los materiales de los cables conductores, el método de montaje, la temperatura ambiente y la forma del termistor.
* **Rango de resistencia:** define los valores de resistencia mínimo y máximo a la temperatura de referencia.
* **Temperatura de transición:** es el punto de Curie en el que la curva R – T del termistor PTC comienza a aumentar bruscamente. Los fabricantes de PTC a menudo definen esta temperatura como el punto en el que existe una relación especificada entre la resistencia mínima (oa una resistencia de potencia cero de 25 ° C) y la resistencia a la temperatura de transición. Por ejemplo, Thermometrics Inc. especifica el punto donde la resistencia es el doble (2 ×) del valor mínimo, mientras que otros fabricantes pueden usar 10 veces (10 ×) el mínimo.
Las principales ventajas de los termistores para la medición de temperatura son: (1) Sensibilidad extremadamente alta. Por ejemplo, un termistor de 2252 Ω tiene una sensibilidad de −100 Ω ⋅ ° C − 1 a temperatura ambiente. Los termistores de mayor resistencia pueden exhibir una sensibilidad de −10 kΩ ⋅ ° C − 1 o más. En comparación, un RTD de platino de 100 Ω tiene una sensibilidad de solo 0,4 Ω ⋅ ° C − 1. (2) Respuesta muy rápida a los cambios de temperatura. (3) Resistencia relativamente alta. Los termistores están disponibles con resistencias de base (a 25 ° C) que van desde cientos hasta millones de ohmios. Esta alta resistencia disminuye el efecto de los cables conductores que pueden causar errores importantes con dispositivos de baja resistencia como los RTD. La alta resistencia y alta sensibilidad de los termistores hacen que sus circuitos de medición y acondicionamiento de señales sean mucho más simples. No se necesitan configuraciones especiales de puente de Wheatstone, de tres o cuatro cables, aunque el uso de un puente de Wheatstone puede mejorar la linealidad de los termistores.

Las principales desventajas de los termistores son su alta no linealidad y su rango de temperatura limitado (normalmente por debajo de 300 ° C). La figura 2.15 muestra la curva R – T para un termistor de 2252 Ω. La curva de un RTD de 100 Ω también se muestra a modo de comparación.

Los termistores tienen dos terminales no polarizados. Según el método por el cual estos terminales se unen al cuerpo cerámico, los termistores se clasifican en tipos de contacto de superficie metalizada y de perlas. Un termistor de tipo perla tiene cables conductores de aleación de platino (de aproximadamente 0,5 a 5 mm de diámetro) que se sinterizan directamente en el cuerpo de cerámica.

El tipo metalizado tiene contactos de superficie metalizados (con o sin cables radiales o axiales) para montaje en superficie o por resorte. Cada tipo puede caracterizarse además por diferencias en geometría, empaque y / o técnicas de procesamiento como se muestra en la Tabla 2.8. La figura 2.16 ilustra el símbolo del circuito y las formas más comunes de termistores.
Los termistores a menudo se montan en tubos de acero inoxidable para protegerlos de entornos hostiles durante su funcionamiento. Normalmente se utiliza grasa termoconductora o sellador de silicona para mejorar el contacto térmico entre el sensor y el tubo (consulte la Figura 2.17). Para mantener la tolerancia de temperatura entre ± 0,05 ° C y ± 1 ° C, los termistores se recortan con láser durante el proceso de fabricación. Los termistores son mecánicamente simples y fuertes, lo que proporciona la base para un sensor de alta confiabilidad. Están disponibles en una amplia gama de tamaños, valores de resistencia base y curvas R – T.
La fabricación de termistores NTC utiliza tecnología cerámica básica: una mezcla de dos o más polvos de óxidos metálicos combinados con aglutinantes adecuados y formados en la forma deseada, secados y sinterizados a una temperatura elevada. Variando los tipos de óxidos utilizados, la temperatura de sinterización y la atmósfera, se puede obtener una amplia gama de resistencias y coeficientes de temperatura.
